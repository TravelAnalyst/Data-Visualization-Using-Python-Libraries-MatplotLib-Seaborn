## Task 6: Create a scatter plot to visualize the relationship between area and price of California real estate. 

```python
scatter_data = pd.read_csv('scatter_data.csv')
scatter_data
```

```python
plt.figure(figsize= (12,8))
scatter = plt.scatter(scatter_data['Area (ft.)'],
            scatter_data['Price'],
            alpha = 0.6,
            c = scatter_data['Building Type'],
            cmap = 'viridis')
plt.legend (*scatter.legend_elements(), loc = 'upper left',
           title = 'Building Type')
plt.title('Relationship between Area and Price of California Real Estate ',
         fontsize = 14, weight = 'bold')
plt.xlabel('Area (ft.)', weight = 'bold')
plt.ylabel("Price (000's of $)")
plt.show()
```

## Using seaborn library to make the same scatterplot.

```python
sns.set_style('darkgrid')
plt.figure(figsize=(12,8))
sns.scatterplot(scatter_data['Price'],
               scatter_data['Area (ft.)'],
               hue = scatter_data['Building Type'],
               palette = ['black', 'darkblue', 'purple', 'pink', 'white'],
               s = 100)
plt.title('Relationship between Area and Price of California Real Estate ',
         fontsize = 14, weight = 'bold')
plt.xlabel('Area (ft.)', weight = 'bold')
plt.ylabel("Price (000's of $)")
plt.show()
```

## Task 7: Create a regression plot to analyze the relationship between advertising budget and sales. Use regplot() as well as lmplot()

```python
scatter2 = pd.read_csv('scatter_plot.csv')
scatter2
```

### Plotting Regression Model using regplot()

```python
#plt.figure(figsize=(10,8))
sns.set(rc = {'figure.figsize':(9,6)})
sns.regplot(x = 'Budget', y = 'Sales',
           data = scatter2,
           scatter_kws={'color':'k'},
           line_kws = {'color': 'r'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Add Expenditure on Sales", fontsize = 16, weight = 'bold')
plt.show()
```

### Plotting Regression Model using lmplot()

```python
sns.lmplot(x = 'Budget',
          y = 'Sales',
          data = scatter2,
          height = 8,
          scatter_kws={'color':'r'},
          line_kws = {'color': 'b'}
          )
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Add Expenditure on Sales", fontsize = 16, weight = 'bold')
plt.show()
```

## Final Task: Create a visualization to show the trend of Python users over the years based on KD Nuggets survey data. Use a bar chart to show the number of participants and a line chart to show the percentage of Python users. Use a twin axis to display the line chart. Set the color and styling of the charts using the Seaborn library. Add appropriate labels and title to the chart. 

```python
bar_line_chart = pd.read_csv('bar_line_chart_data.csv')
bar_line_chart
```

```python
from matplotlib.ticker import PercentFormatter
```

```python
sns.set_style('white')
fig, ax = plt.subplots(figsize = (10,7))
ax.bar(bar_line_chart['Year'],
      bar_line_chart['Participants'],
      color = 'k')
ax.set_ylabel("number of Participants",
             weight = 'bold')
ax.tick_params (axis = 'y',
               width = 2,
               labelsize = 'large')
ax1 = ax.twinx()
ax1.set_ylim(0,1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
ax1.plot(bar_line_chart['Year'],
        bar_line_chart['Python Users'],
        color = '#b60000',
        marker = 'D')
ax1.set_ylabel('Python Users',
              color = "#b60000",
              weight = 'bold')
ax.tick_params(axis = 'y',
              colors = '#b60000',
              width = 2,
              labelsize = 'large')
ax.set_title('KD Nuggets Survey Python Users (2012 - 2019)', fontsize = '14', weight = 'bold')
```

## Interpretation 

This plot shows the number of survey participants and the percentage of participants who reported using Python as their primary language from 2012 to 2019. The black bars represent the total number of participants, while the red diamond markers represent the percentage of participants who reported using Python.

From the plot, we can see that the number of survey participants has steadily increased over the years. Additionally, the percentage of participants using Python has also steadily increased, with a sharp rise between 2016 and 2017. In 2019, Python usage surpassed 50% of all survey respondents.

Overall, this plot suggests that Python is becoming an increasingly popular programming language among data professionals and is likely to continue to gain popularity in the future.
